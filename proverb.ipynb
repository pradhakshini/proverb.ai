```python
from transformers import pipeline, set_seed
import gradio as gr

# Load the GPT-2 text generation model
generator = pipeline("text-generation", model="gpt2")
set_seed(42)

# Function to generate text based on custom prompt
def generate_proverb_with_prompt(prompt):
    if not prompt.strip():
        prompt = "A wise man once said"
    result = generator(prompt, max_length=30, num_return_sequences=1)
    return result[0]["generated_text"].strip()
```

```python
# Gradio interface with input textbox
demo = gr.Interface(
    fn=generate_proverb_with_prompt,
    inputs=gr.Textbox(lines=2, placeholder="Enter a custom prompt like 'Life is...'"),
    outputs="text",
    title="✨ AI Proverb Generator ✨",
    description="Type your own prompt or use something like 'A wise man once said', then click Submit."
)

# Launch the app
demo.launch()
```
